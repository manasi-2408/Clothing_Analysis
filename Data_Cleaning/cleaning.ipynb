## Cleaning the data

### Date Cleaning

```python
def parse_mixed_date(x):
    try:
        return pd.to_datetime(x, format='%d-%m-%Y')
    except:
        try:
            return pd.to_datetime(x, format='%m-%d-%Y')
        except:
            return pd.NaT

data['PurchaseDate'] = data['PurchaseDate'].astype(str).apply(parse_mixed_date)

data['PurchaseDate'] = data['PurchaseDate'].dt.strftime('%d-%m-%Y')
```

```python
data['Purchasedate'] = pd.to_datetime(data['PurchaseDate'], format='%d-%m-%Y')
```

### City Cleaning

```python
data['City'] = data['City'].fillna(data['City'].mode()[0])
```

### UnitPrice Cleaning

```python
data['UnitPrice'] = data['UnitPrice'].fillna(data['UnitPrice'].median())
```

### Age Cleaning

```python
data['Age'] = data['Age'].fillna(data['Age'].median())
```

### Gender Cleaning

```python
data['Gender'] = data['Gender'].replace({'F':'Female', 'M':'Male'})
```

```python
for i in range(len(data)):
    if data.loc[i, 'Category'] == "Women's Wear" or \
       'dress' in data.loc[i, 'Product'].lower() or \
       'skirt' in data.loc[i, 'Product'].lower():
        data.loc[i, 'Gender'] = "Female"
```

```python
data['Gender'] = data['Gender'].fillna(data['Gender'].mode()[0])
```

## Saved the data

```python
data.to_csv(r'cleaned_excel.csv', index=False)
```
