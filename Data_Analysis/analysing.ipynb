## New Columns to be made

```python
data['Age_Bin'] = pd.cut(data['Age'], bins=[12, 19, 35, 59, 85], labels=['Teenagers', 'Adults', 'Seniors', 'Old People'], right=True, include_lowest=True)
```

```python
def age_group(age):
    if age < 30:
        return "18-29"
    elif age < 45:
        return "30-44"
    elif age < 60:
        return "45-59"
    else:
        return "60+"

data["Age_Group"] = data["Age"].apply(age_group)
```

```python
data["Total_Spend"] = data["Quantity"] * data["UnitPrice"]
```

```python
data["Purchase_Year"] = pd.to_datetime(data["PurchaseDate"], dayfirst=True).dt.year
data["Purchase_Month"] = pd.to_datetime(data["PurchaseDate"], dayfirst=True).dt.month
```

```python
def get_season(month):
    if month in [12, 1, 2]:
        return "Winter"
    elif month in [3, 4, 5]:
        return "Summer"
    elif month in [6, 7, 8]:
        return "Monsoon"
    else:
        return "Autumn"

data["Season"] = data["Purchase_Month"].apply(get_season)
```

```python
cluster_map = {0: "Frequent Buyers", 1: "Premium Buyers", 2: "Budget Buyers"}
data["Cluster_Label_Desc"] = data["Clusters"].map(cluster_map)
```

## Saving the data
